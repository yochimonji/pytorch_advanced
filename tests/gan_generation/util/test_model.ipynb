```python
import os
import sys
sys.path.append(os.path.abspath("../../../"))
from typing import cast

import torch
import matplotlib.pyplot as plt
%matplotlib inline

from pytorch_advanced.gan_generation.util.model import Generator, Discriminator
```

```python
g = Generator(z_dim=20, image_size=64)
input_z = torch.rand(1, 20, 1, 1)
fake_images = g(input_z)
```

```python
img_transformed = cast(torch.Tensor, fake_images[0][0]).detach().numpy()
plt.imshow(img_transformed, cmap="gray")
plt.show()
```

```python
d = Discriminator(image_size=64)
out = d(fake_images)
print(torch.nn.Sigmoid()(out))
```
